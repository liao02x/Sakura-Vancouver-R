```python
import requests
import json
import pandas as pd
```

```python
url = 'http://maps.vcbf.ca/service/poi/getpoiformaps.php'
head = ['Cultivar',
        'Neighborhood',
        'Location',
        'Blooming',
        'Longitude',
        'Latitude',
        'id',
        'Description',
        'ForumLink',
        'PicLink',
        'Favorate'
       ]

r = requests.get(url)
```

```python
json_data = json.loads(r.text)
```

```python
data = pd.DataFrame(json_data['data'], columns=head)
```

```python
data.head()
```

```python
data.shape
```

```python
data.dropna(subset=['Blooming'], inplace=True)
```

```python
data.shape
```

```python
data['Start'] = data['Blooming'].map(lambda x: x.split('-')[0])
data['End'] = data['Blooming'].map(lambda x: x.split('-')[1])
```

```python
data.head()
```

```python
data.to_csv('sakura.csv')
```
